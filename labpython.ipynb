```python
import random
number = []
number1 = []
number2 = []

#function1-update list
#function2-sum list
#function3-find avg
#function4-find highest and lowest
#functin5-find num smaller than avg

def update(mylist):
  for x in range(1,11):
    mylist.append(random.randrange(20,50))

def getsum(mylist):
  return sum(mylist)

def getavg(sumval,length):
  avg=sumval/length
  return avg

def getmax(mylist):
  return max(mylist)

def getmin(mylist):
  return min(mylist)

def getsmallvalue(avgval,mylist):
  #for y in mylist:
    #if (y<avgval):
      #return y
  return [y for y in mylist if y <avgval]

print(number)
update(number)
total=getsum(number)
average=getavg(total,len(number))
print("List:" ,number)
print("Total sum:" ,total)
print("Average:" ,average)
print("Maximum:" ,getmax(number))
print("Minimum:" ,getmin(number))
print("Small value:" ,getsmallvalue(average,number))

print("\n")

print(number1)
update(number1)
total1=getsum(number1)
average1=getavg(total1,len(number1))
print("List:" ,number1)
print("Total sum:" ,total1)
print("Average:" ,average1)
print("Maximum:" ,getmax(number1))
print("Minimum:" ,getmin(number1))
print("Small value:" ,getsmallvalue(average1,number1))

print("\n")

print(number2)
update(number2)
total2=getsum(number2)
average2=getavg(total2,len(number2))
print("List:" ,number2)
print("Total sum:" ,total2)
print("Average:" ,average2)
print("Maximum:" ,getmax(number2))
print("Minimum:" ,getmin(number2))
print("Small value:" ,getsmallvalue(average2,number2))

print("\n")

total=sum(number)
print(total)

average=total/ len(number)
print("The average is", average)

maxi =max(number)
print("Maximum: ", maxi)

mini =min(number)
print("Minimum: ", mini)

for i in number:
  if (i < average):
    print("Value less than average",i)

for i in number:
  if (i > average):
    print("Value bigger than average",i)
```
